### Draw pie charts to visualize the data

```python
shapes = ['WEATHER_DELAY', 'CARRIER_DELAY', 'NAS_DELAY', 'SECURITY_DELAY', 'LATE_AIRCRAFT_DELAY']
values = [ 7721, 47144, 61801, 332, 51425]

s = pd.Series(values, index=shapes)
s
```

```python
from matplotlib import font_manager as fm
from  matplotlib import cm

labels = s.index
sizes = s.values
explode = (0.1,0.01,0.01,0.2,0.01)  # "explode" ， show the selected slice

fig, axes = plt.subplots(figsize=(15,9),ncols=2) # Set the drawing area size
ax1, ax2 = axes.ravel()

colors = cm.rainbow(np.arange(len(sizes))/len(sizes)) # colormaps: Paired, autumn, rainbow, gray,spring,Darks
patches, texts, autotexts = ax1.pie(sizes, labels=labels, autopct='%1.1f%%',explode=explode,
        shadow=False, startangle=170, colors=colors, labeldistance=1.2,pctdistance=0.80, radius=0.4)
# labeldistance: Controls where the label is displayed
# pctdistance: Controls the % display position
# radius: Control the distance of section protrusion

ax1.axis('equal')  

# Reset the font size
proptease = fm.FontProperties()
proptease.set_size('xx-large')
# font size include: ‘xx-small’,x-small’,'small’,'medium’,‘large’,‘x-large’,‘xx-large’ or number, e.g. '12'
plt.setp(autotexts, fontproperties=proptease)
plt.setp(texts, fontproperties=proptease)

ax1.set_title('Percentage of aircraft delays of various causes', loc='center',fontsize=17,bbox={'facecolor':'8', 'pad':5})

# ax2 show （legend）
ax2.axis('off')
ax2.legend(patches, labels, loc='upper left',fontsize=15)

plt.tight_layout()
# plt.savefig("pie_shape_ufo.png", bbox_inches='tight')
plt.savefig('Demo_project_final.jpg')
plt.show()
```

From the pie chart we can understand the number of Nas Delay is highest, while the impact of weather and security is pretty low.Further analysis, we can infer that the United States has extreme weather conditions, relatively large traffic volume and traffic control, which leads to the increase of NAS DELAY.Carrier Dlay also has a high number of flights in the United States due to unpunctuality of customers, inadequate crew Numbers, strikes and airport maintenance.It is because the large population of the United States, and aircraft as a common way for americans to travel, American airports have a lot of flights.
```python
def OP_UNIQUE_CARRIER(group):
    return {'min': group.min(), 'max': group.max(),
            'count': group.count(), 'mean': group.mean()}
flight_operator = data['ARR_DEL15'].groupby(data['OP_UNIQUE_CARRIER']).apply(OP_UNIQUE_CARRIER).unstack()
flight_operator=flight_operator.sort_values('count',ascending=False)
flight_operator
```

```python
import matplotlib.patches as mpatches
import seaborn as sns
delay_type = lambda x:((0,1)[x > 5],2)[x > 45]
data['DELAY_LEVEL'] = data['DEP_DELAY'].apply(delay_type)
fig = plt.figure(1, figsize=(10,7))
ax = sns.countplot(y='OP_UNIQUE_CARRIER', hue='DELAY_LEVEL', data=data)

plt.setp(ax.get_xticklabels(), fontsize=12, weight = 'normal', rotation = 0);
plt.setp(ax.get_yticklabels(), fontsize=12, weight = 'bold', rotation = 0);
ax.yaxis.label.set_visible(False)
plt.xlabel('Flight count', fontsize=16, weight = 'bold', labelpad=10)

L = plt.legend(loc='lower right')
L.get_texts()[0].set_text('on time (t < 5 min)')
L.get_texts()[1].set_text('small delay (5 < t < 45 min)')
L.get_texts()[2].set_text('large delay (t > 45 min)')
plt.grid()
plt.title('Airline Delay',fontsize=20)
plt.show()
```

This figure shows the number of delays less than 15 minutes, followed by delays in the range of 5 < t < 45 minutes, and finally, the route delay is greater than 45 minutes.Our guess is that, unrelated to the airline, delays of more than 45 minutes are in the minority.I'll delve into the number of delays per flight in the next section.We can also see that WN airline has a relatively low proportion of delays greater than 45 minutes

```python
#This is a function I googled to plot the heat map.
def heatplot(plotmat, names=None, names1=None,cmap="Greys", cmap_range=None, ax=None, **kwargs):
   
    if ax is None:
        ax = plt.gca()

    nvars = len(plotmat)

    if cmap_range is None:
        vmax = np.nanmax(plotmat) * 1.15
        vmin = np.nanmin(plotmat) * 1.15
    elif len(cmap_range) == 2:
        vmin, vmax = cmap_range
    else:
        raise ValueError("cmap_range argument not understood")

    mat_img = ax.matshow(plotmat, cmap=cmap, vmin=vmin, vmax=vmax, **kwargs)

    plt.colorbar(mat_img, shrink=.75)
 
    ax.xaxis.set_ticks_position("bottom")
    ax.set_xticklabels(names, rotation=90)
    ax.set_yticklabels(names1)

    minor_ticks = np.linspace(-.5, nvars - 1.5, nvars)
    ax.set_xticks(minor_ticks, True)
    ax.set_yticks(minor_ticks, True)
    major_ticks = np.linspace(0, nvars - 1, nvars)
    ax.set_xticks(major_ticks)
    ax.set_yticks(major_ticks)
    ax.grid(False, which="major")
    ax.grid(True, which="minor", linestyle="-")

    return ax
```
